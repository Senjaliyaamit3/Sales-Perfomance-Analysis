```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("sales_big.csv")

print("First 5 Rows:")
print(df.head())

#Basic Operations
# Total Sales
print("\nTotal Sales:", df["Sales"].sum())

# Average Sales
print("Average Sales:", round(df["Sales"].mean(),2))

# Max and Min Sales
print("Max Sale:", df["Sales"].max())
print("Min Sale:", df["Sales"].min())

# Filtering (No groupby)

# Only Laptop Sales
laptop_sales = df[df["Product"] == "Laptop"]
print("\nLaptop Total Sales:", laptop_sales["Sales"].sum())

# Only North Region
north_sales = df[df["Region"] == "North"]
print("North Region Total:", north_sales["Sales"].sum())

# Unique Values
print("\nUnique Regions:", df["Region"].unique())
print("Unique Products:", df["Product"].unique())

# Sorting
top_sales = df.sort_values(by="Sales", ascending=False)
print("\n Top 5 Sales:\n", top_sales.head())

# Visualization
# Sales Trend
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")

plt.plot(df["Date"], df["Sales"], marker='o')
plt.title("Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Product-wise total sales (without groupby)
laptop = df[df["Product"] == "Laptop"]["Sales"].sum()
mobile = df[df["Product"] == "Mobile"]["Sales"].sum()
tablet = df[df["Product"] == "Tablet"]["Sales"].sum()

sales_data = [laptop, mobile, tablet]
labels = ["Laptop", "Mobile", "Tablet"]

plt.pie(sales_data, labels=labels, autopct="%1.1f%%")
plt.title("Sales Distribution by Product")

plt.show()
```
